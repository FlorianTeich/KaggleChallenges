```python
# Connect to Postgres
import os
import sqlalchemy
import pandas as pd

engine = sqlalchemy.create_engine(
            "postgresql+psycopg2://"
            + os.getenv("SQL_USER")
            + ":"
            + os.getenv("SQL_PASSWORD")
            + "@"
            + os.getenv("SQL_HOST")
            + ":"
            + os.getenv("SQL_PORT")
            + "/"
            + os.getenv("SQL_DB")
        )
titanic_data = pd.read_sql_table("titanic", engine)
titanic_data
```

```python
# Connect to PySpark
import pyspark
from pyspark.sql import SparkSession

spark = SparkSession \
    .builder \
    .appName("TitanicExample") \
    .config("spark.jars", "/srv/KaggleChallenge/plugins/postgresql-42.5.1.jar") \
    .getOrCreate()

df = spark.read \
    .format("jdbc") \
    .option("url", "jdbc:postgresql://" + os.getenv("SQL_HOST") +":" + os.getenv("SQL_PORT") + "/" + os.getenv("SQL_DB")) \
    .option("dbtable", "titanic") \
    .option("user", os.getenv("SQL_USER")) \
    .option("password", os.getenv("SQL_PASSWORD")) \
    .option("driver", "org.postgresql.Driver") \
    .load()

df.printSchema()
```

```python
# Connect to mlflow
import mlflow

mlflow.set_tracking_uri("http://mlflow:5000")
mlflow.set_experiment("my-experiment")
```

```python
#import logging
#logging.basicConfig(level=logging.DEBUG)
from pyspark.sql import functions as F

df = df.withColumn("mod10_split", F.monotonically_increasing_id() % 10)
```

```python
from pyspark.ml.feature import StringIndexer
from pyspark.ml import Pipeline
from pyspark.ml.feature import HashingTF, Tokenizer
from pyspark.ml.feature import VectorAssembler

# Split Dataset
df_train = df.filter(F.col("mod10_split").isin([0, 1, 2, 3, 4, 5]))
df_val = df.filter(F.col("mod10_split").isin([6, 7]))
df_test = df.filter(F.col("mod10_split").isin([8, 9]))

sex_si = StringIndexer(inputCol="sex", outputCol="sex_category", stringOrderType="frequencyDesc", handleInvalid="keep")
sex_si_model = sex_si.fit(df)

vecAssembler = VectorAssembler(outputCol="features", handleInvalid="keep")
vecAssembler.setInputCols(["age", "fare", "body", "sex_category"])

pipeline = Pipeline(stages=[sex_si_model, vecAssembler])

# Fit the pipeline to training documents.
model = pipeline.fit(df)

# Make predictions on test documents and print columns of interest.
prediction = model.transform(df)
```
